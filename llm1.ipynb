```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
```

```python
model_name = "gpt2"  # small, fast model

# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
generator = pipeline("text-generation", model=model, tokenizer=tokenizer)
```

```python
prompt = "Once upon a time, in a jungle,"
result = generator(prompt, max_length=50, num_return_sequences=1)

print("Generated Text:")
print(result[0]['generated_text'])
```

```python
prompts = [
    "The future of AI is",
    "A cheetah is the fastest animal because",
    "Once upon a time, a young girl found a magical"
]

for prompt in prompts:
    result = generator(prompt, max_length=40, num_return_sequences=1)
    print(f"Prompt: {prompt}")
    print("Output:", result[0]['generated_text'])
    print("-" * 50)
```
